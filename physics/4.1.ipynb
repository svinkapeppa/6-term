```python
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
p = 750 - np.array([730, 710, 690, 670, 650, 630,
                    610, 590, 570, 550, 530, 510,
                    490, 470, 450, 430, 400])
N = np.array([3301, 2998, 2698, 2440, 2116,
              1808, 1467, 1032, 780, 558,
              259, 148, 99, 64, 15, 14, 13])
```

```python
x2 = np.arange(260)
y2 = (N[11] - N[5]) / (p[11] - p[5]) * x2 + (p[11] * N[5] - p[5] * N[11]) / (p[11] - p[5])
```

```python
plt.figure(figsize=(12,8))

plt.plot(p, N, label=' ')
plt.plot(x2, y2)
plt.hlines(0, 0, 350)

plt.xlabel(r'$P, $', horizontalalignment='right',
           fontsize='x-large', x=1)
h = plt.ylabel(r'$N$', verticalalignment='top',
               fontsize='x-large', horizontalalignment='right', y=1)
h.set_rotation(0)

plt.title(r' $N = N(P)$')
plt.grid(ls='--')
plt.legend()
plt.savefig("N(P)")
plt.show()
```
